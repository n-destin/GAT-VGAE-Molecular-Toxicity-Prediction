## Data Splitting
Quick file to create the fivefold split of the Clintox and Tox21 datasets, with 0.8/0.1/0.1 train/test/val split.

```python
import pandas as pd

from random import shuffle, sample
```

```python
from google.colab import drive
drive.mount('/content/drive')

df_clintox = pd.read_csv("/content/drive/Shareddrives/cs_89_network_science/data/clintox.csv")
df_tox21 = pd.read_csv("/content/drive/Shareddrives/cs_89_network_science/data/tox21.csv")
```

```python
def generate_split(df, train=0.8, val=0.1):
  n = len(df)

  idx = list(range(n))
  shuffle(idx)
  train_idx = idx[:int(train*n)]
  test_idx = idx[int(train*n):int((1-val)*n)]
  val_idx = idx[int((1-val)*n):]

  df['train'] = df['test'] = df['val'] = 0
  df.loc[train_idx, 'train'] = 1
  df.loc[test_idx, 'test'] = 1
  df.loc[val_idx, 'val'] = 1

def fivefold_split(df, name):
  n = len(df)

  idx = list(range(n))
  shuffle(idx)
  segs = [idx[i*int(n/5):(i+1)*int(n/5)] for i in range(4)] + [idx[4*int(n/5):]]

  i = 1
  for seg_i in segs:
    df['train'] = df['test'] = df['val'] = 0
    for seg_j in segs:
      if seg_i != seg_j: # set 4 of the 5 folds to training data
        df.loc[seg_j, 'train'] = 1
      else: # split fifth between testing and validation data
        test_idx = seg_j[:int(len(seg_j)/2)]
        val_idx = seg_j[int(len(seg_j)/2):]
        df.loc[test_idx, 'test'] = 1
        df.loc[val_idx, 'val'] = 1
    df.to_csv("/content/drive/Shareddrives/cs_89_network_science/data/" + name + "_fivefoldsplit#" +str(i) + ".csv", index=False)
    i += 1
```

```python
fivefold_split(df_clintox, "clintox")
fivefold_split(df_tox21, "tox21")
```
